```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

定義されていない確率分布の最尤推定を実施する場合には，最適化ツールを使う

```python
def nLL(theta):
    return len(mydata) / 2 * np.log(2 * np.pi * theta[1]**2) \
           + np.sum((mydata - theta[0])**2) / 2 / theta[1]**2
```

```python
from scipy.optimize import minimize, Bounds

theta0 = np.array([0.0, 1.0])  # 初期値（推定されるパラメータの予測値）
bounds = Bounds([-np.inf, 1e-6], [np.inf, np.inf])  # パラメータの範囲
res = minimize(nLL, theta0, bounds=bounds)
print('estimated:', res['x'])
```

```python
n_list = [100,316,1000,3162,10000,31622,100000]
mu1 = 10
mu2 = 15
sum_mu = mu1+mu2
var1 = 3.0
var2 = 6.0
sum_var = np.sqrt(var1*var1+var2*var2)
means = []
sigmas = []
for n in n_list:
    rvs1 = stats.norm.rvs(loc=mu1, scale=var1, size=n)
    rvs2 = stats.norm.rvs(loc=mu2, scale=var2, size=n)
    sum_rvs = rvs1+rvs2
    #print(stats.norm.fit(sum_rvs))
    #print(sum_var)
    mean,sigma = stats.norm.fit(sum_rvs)
    means.append(mean)
    sigmas.append(sigma)
```

```python
# === グラフの作成 ===
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

# 平均の比較
axes[0].plot(n_list, means, 'o-', label='sample_mean')
axes[0].axhline(sum_mu, color='r', linestyle='--', label='expectation')
axes[0].set_xscale('log')
axes[0].set_xlabel("n")
axes[0].set_ylabel("mean")
axes[0].set_title("mean")
axes[0].legend()
axes[0].grid(True)

# 標準偏差の比較
axes[1].plot(n_list, sigmas, 'o-', label='sample standard deviation')
axes[1].axhline(sum_var, color='r', linestyle='--', label='expected standard deviation')
axes[1].set_xscale('log')
axes[1].set_xlabel("n")
axes[1].set_ylabel("standard deviation")
axes[1].set_title("standard deviation")
axes[1].legend()
axes[1].grid(True)

plt.tight_layout()
plt.savefig("kadai2.png", dpi=300, bbox_inches='tight')
plt.show()
```

## 混合分布

```python
import numpy as np

def myrvs(size : int):
    result = np.zeros(size)
    for i in range(size):
        u = np.random.rand()
        if u < 0.2:
            n = -4.0 + 0.5 * np.random.randn()
        elif u < 0.5:
            n = 0.0 + 1.0 * np.random.randn()
        else:
            n = 3.0 + 2.0 * np.random.randn()
        result[i] = n
    return result
```

```python
import matplotlib.pyplot as plt
mydata = myrvs(1000)
plt.hist(mydata, bins=100)
```

成分分布が正規分布の場合には，sklearn.mixture.GaussianMixtureを使えばEMアルゴリズムの実行が可能

```python
from sklearn.mixture import GaussianMixture
mydata2d = mydata.reshape((-1, 1))
gm = GaussianMixture(n_components=3).fit(mydata2d)
```

```python
for i in range(len(gm.means_)):
    mean_ = gm.means_[i]
    cov_ = gm.covariances_[i]
    print("Component ", i+1, ": mean vector = ", mean_, ", covariance matrix = ", cov_)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture
from scipy.stats import multivariate_normal


mean1 = np.array([2, 3])
cov1 = np.array([[3, 1], [1, 2]])
mean2 = np.array([7, 5])
cov2 = np.array([[2, -0.5], [-0.5, 1]])
mean3 = np.array([4, 8])
cov3 = np.array([[1, 0.3], [0.3, 1.5]])
        
def myrvs(siz : int):
    data1 = multivariate_normal(mean1, cov1).rvs(size=siz*2)
    data2 = multivariate_normal(mean2, cov2).rvs(size=siz*3)
    data3 = multivariate_normal(mean3, cov3).rvs(size=siz*5)

    result = np.empty((siz*10, 2))
    result[0:siz*2] = data1
    result[siz*2:siz*5] = data2
    result[siz*5:siz*10] = data3 
    return result
    
n_list = [10,100,1000]
graph_mean = []
graph_cov = []
graph_prob = []
for n in n_list:
    print("----------------------------------------------------")
    mydata2d = myrvs(n)
    gm = GaussianMixture(n_components=3).fit(mydata2d)

        
    true_means = np.array([mean1, mean2, mean3])  
    estimated_means = gm.means_                  
    
    dist_matrix = np.linalg.norm(estimated_means[:, None, :] - true_means[None, :, :], axis=2)
    indices = np.argmin(dist_matrix, axis=1)
    sorted_means = estimated_means[np.argsort(indices)]
    sorted_covs = gm.covariances_[np.argsort(indices)]
    sorted_probs = gm.weights_[np.argsort(indices)]

    for i in range(len(gm.means_)):
        mean_ = sorted_means[i]
        cov_ = sorted_covs[i]
        prob_ = sorted_probs[i]
        print("Component ", i+1, ": mean vector = ", mean_, ", covariance matrix = ", cov_, ", prob = ", prob_)
    graph_mean.append(sorted_means)
    graph_cov.append(sorted_covs)
    graph_prob.append(sorted_probs)
```

```python
graph_mean = np.array(graph_mean)
graph_cov = np.array(graph_cov)
graph_prob = np.array(graph_prob)
plt.figure(figsize=(6,6))
true_means_list = [mean1, mean2, mean3]
true_covs = [cov1, cov2, cov3]
true_probs = [0.2,0.3,0.5]
#--------平均ベクトル-----------
for i in range(3): 
    x_vals = graph_mean[:, i, 0] 
    y_vals = graph_mean[:, i, 1] 
    
    plt.scatter(x_vals, y_vals, label=f'mean vactor{i+1}')
    
    for j in range(len(x_vals)-1):
        plt.annotate('', xy=(x_vals[j+1], y_vals[j+1]), xytext=(x_vals[j], y_vals[j]),
                     arrowprops=dict(arrowstyle='->', lw=1.5))

    plt.scatter(true_means_list[i][0], true_means_list[i][1],
             marker='*', s=150, label='expectation')

    plt.xlabel('mean of x')
    plt.ylabel('mean of y')
    plt.title('GMM Estimated Mean')
    plt.grid(True)
    plt.legend()
    plt.savefig(f"kadai3vec{i+1}.png", dpi=300, bbox_inches='tight')
    plt.show()
#------------共分散行列---------------
elements = ['var_x', 'var_y', 'cov_xy']
colors = ['r', 'g', 'b']

for i in range(3):
    plt.figure(figsize=(6,4))
    for j, elem in enumerate(elements):
        if elem == 'var_x':
            y_vals = graph_cov[:, i, 0, 0]
            true_val = true_covs[i][0, 0]
        elif elem == 'var_y':
            y_vals = graph_cov[:, i, 1, 1]
            true_val = true_covs[i][1, 1]
        elif elem == 'cov_xy':
            y_vals = graph_cov[:, i, 0, 1]
            true_val = true_covs[i][0, 1]
        
        plt.plot(n_list, y_vals, marker='o', label=f'{elem} (estimated)')
        plt.axhline(true_val, linestyle='--', color='k', alpha=0.6, label=f'{elem} (true)' if j==0 else "")
    
    plt.xscale('log')
    plt.xlabel('n')
    plt.ylabel('covariance')
    plt.title(f'Component {i+1} Covariance')
    plt.grid(True)
    plt.legend()     
    plt.savefig(f"kadai3cov{i+1}.png", dpi=300, bbox_inches='tight')
    plt.show()
#---------確率--------
for i in range(3): 
    p_vals = graph_prob[:, i] 
    plt.plot(n_list, p_vals, 'o-', label=f'weight{i+1}')
    plt.axhline(true_probs[i], color='r', linestyle='--', label='true wight')

plt.xlabel('n')
plt.xscale('log')
plt.ylabel('weight')
plt.title('GMM Estimated weight')
plt.grid(True)
plt.legend()
plt.savefig("kadai3p.png", dpi=300, bbox_inches='tight')
plt.show()
```

## カーネル密度推定

sklearn.neighbors.KernelDensity を用いれば，いくつかの用意されたカーネルに対する密度推定を実施することができる．

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KernelDensity
```

混合正規分布の項で作成した乱数生成器を使用

```python
def myrvs(size : int):
    result = np.zeros(size)
    for i in range(size):
        u = np.random.rand()
        if u < 0.2:
            n = -4.0 + 0.5 * np.random.randn()
        elif u < 0.5:
            n = 0.0 + 1.0 * np.random.randn()
        else:
            n = 3.0 + 2.0 * np.random.randn()
        result[i] = n
    return result
```

```python
mydata = myrvs(10000)
```

３種類のカーネルを用いて密度推定
* バンド幅を変えると（大きくすると）カーネル毎の差が顕著に可視化される．

```python
kde_gaussian = KernelDensity(bandwidth=0.1, kernel='gaussian').fit(mydata.reshape((-1, 1)))
kde_tophat = KernelDensity(bandwidth=0.1, kernel='tophat').fit(mydata.reshape((-1, 1)))
kde_linear = KernelDensity(bandwidth=0.1, kernel='linear').fit(mydata.reshape((-1, 1)))
```

```python
fig = plt.figure()
x_array = np.linspace(-10, 10, num=1000)
p_gaussian = np.exp(kde_gaussian.score_samples(x_array.reshape((-1, 1))))
p_tophat = np.exp(kde_tophat.score_samples(x_array.reshape((-1, 1))))
p_linear = np.exp(kde_linear.score_samples(x_array.reshape((-1, 1))))

plt.plot(x_array, p_gaussian, label="gaussian")
plt.plot(x_array, p_tophat, label="tophat")
plt.plot(x_array, p_linear, label="linear")
plt.ylabel('p(x)')
plt.xlabel('x')
plt.legend()
```

```python
from scipy.stats import norm

true_pdf = (0.2 * stats.norm.pdf(x_array, loc=-4, scale=0.5) +
            0.3 * stats.norm.pdf(x_array, loc=0, scale=1) +
            0.5 * stats.norm.pdf(x_array, loc=3, scale=2))

gaussian_squared_diff = (true_pdf - p_gaussian)**2
gaussian_error = np.sum(gaussian_squared_diff)
tophat_squared_diff = (true_pdf - p_tophat)**2
tophat_error = np.sum(tophat_squared_diff)
linear_squared_diff = (true_pdf - p_linear)**2
linear_error = np.sum(linear_squared_diff)

print(f"gaussian error  = {gaussian_error:.6f}")
print(f"tophat error  = {tophat_error:.6f}")
print(f"linear error  = {linear_error:.6f}")
```
