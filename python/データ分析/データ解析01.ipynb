## 制御
その他数多くのプログラミング言語同様，if文やfor文が定義されている．

一方，数多くのプログラミング言語と異なり，インデントが文法的に重要な意味を持つ．例えば，C言語などでは，インデントは読みやすさのために通常挿入するものの，これがなくても文法的には問題ない．他方，Pythonでは，インデントによりブロックを構成します．

```python
for i in range(10):
    print(i)
print("for-loopの外側")
```

```python
mylist = [0, 1, 2.0, "str", []]
for element in mylist:
    print(element, type(element))
```

```python
mylist = [0, 1, 2, "str", [1]]
for element in mylist:
    if element == 1:
        print("１つ目の条件文にマッチ", element, type(element))
    elif element == [1]:
        print("２つ目の条件文にマッチ", element, type(element))
    elif element in [0, 2]:
        print("３つ目の条件文にマッチ", element, type(element))        
    else:
        print("条件文にマッチしない", element, type(element))
```

```python
mylist1 = [0, 1, 2, 3, 4]
mylist2 = ["zero", "one", "two", "three", "four"]
for elem1, elem2 in zip(mylist1, mylist2):
    print(elem1, elem2)
```

```python
mylist1 = [0, 1, 2, 3, 4]
mylist2 = ["zero", "one", "two", "three", "four"]
for i in range(len(mylist1)):
    elem1 = mylist1[i]
    elem2 = mylist2[i]
    print(elem1, elem2)
```

#### 特異値分解
任意の$m \times n$次元実行列 $R$ に対して，以下を満たす行列$(U, S, V)$が存在する．
$$
R = U \cdot S \cdot V^\mathrm{T} \enspace,
$$
ここで，
* $U$: $m$次元直行行列
* $V$: $n$次元直行行列
* $S$: $m \times n$次元対角行列

```python
u, s, vt = np.linalg.svd(R)
```

```python
print(s)
```

```python
print(np.diag(s))
```

```python
np.dot(u, u.T)
```

```python
np.dot(vt, vt.T)
```

```python
print(R)
```

```python
np.dot(np.dot(u, np.diag(s)), vt)
```

```python
u @ np.diag(s) @ vt
```

```python
print(u @ np.diag(s) @ vt - R)
```

#### 逆行列とMoore-Penroseの擬似逆行列
任意の$m \times n$次元実行列 $A$ について，以下を満たす行列$B$が唯一存在する:
* $A B A = A$
* $B A B = B$
* $(A B)^\mathrm{T} = A B$
* $(B A)^\mathrm{T} = B A$

なお，$A$が正則ならば，$B = A^{-1}$である．

```python
inv_R = np.linalg.inv(R)
print(inv_R)
print(inv_R @ R)
```

```python
R[2, :] = 0
print(R)
```

```python
pinv_R = np.linalg.pinv(R)
print(pinv_R)
```

```python
print(np.dot(np.dot(R, pinv_R), R) - R)
```

```python
print(np.dot(np.dot(pinv_R, R), pinv_R) - pinv_R)
```

```python
print(np.dot(R,pinv_R).T-np.dot(R,pinv_R))
```
